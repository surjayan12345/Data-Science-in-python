Uplaod Kaggle json file

```python
import os
```

```python
import shutil
```

```python


# Define the source path of your kaggle.json file
source = 'kaggle.json'  # Update this if the file is in a different location
destination = os.path.expanduser('~/.kaggle/kaggle.json')

# Create the .kaggle directory if it doesn't exist
os.makedirs(os.path.dirname(destination), exist_ok=True)

# Move the kaggle.json file to the .kaggle directory
shutil.move(source, destination)

# Set the permissions for the kaggle.json file (optional, mainly for Unix-based systems)
if os.name != 'nt':
    os.chmod(destination, 0o600)

# Verify the contents of the .kaggle directory
print(os.listdir(os.path.dirname(destination)))
print("kaggle.json configured successfully.")
```
