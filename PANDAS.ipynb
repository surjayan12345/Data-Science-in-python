## Manipulating a DATA FRAME

```python
# adding the column to a dataframe 

california_df['Price'] = California_housing.target #adding the price column from california dataset
```

```python
california_df.head()
```

```python
#removing a row (for row axis = 0 , for column axis = 0 )

#here we will remove the first row and index should be 0

california_df.drop(index = 0 , axis = 0)
```

```python
#drop a column

california_df.drop (columns = 'Longitude' ,  axis = 1)
```

```python
#locating the row using the index value

california_df.iloc[2]
```

```python
#print a specific columns

print(california_df.iloc[: , 0])   #first column
```

```python
print(california_df.iloc[: , 5])   #six column
```

```python
print(california_df.iloc[: , -1])   #last column
```
