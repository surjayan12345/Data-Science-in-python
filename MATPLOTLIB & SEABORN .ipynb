```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
plt.plot(x,z)
plt.xlabel('angle')
plt.ylabel('cos value')
plt.title('cos wave')
plt.show()
```

```python
x  =  np.linspace(10,-10,50)
y = x**2
plt.plot(x,y ,'r' ) #r = red
plt.title('parabola')
```

```python
g = np.linspace(-4 , 4 , 80)
plt.plot(g, np.sin(g) , 'r--')
plt.plot(g, np.cos(g) , 'g--')
plt.show()
```

```python
# BAR CHART


fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['English' , 'French' , 'Spanish' , 'Latin' , 'Green']
people = [100,50,150,40,70]
ax.bar(languages , people)
plt.xlabel('LANGUAGES')
plt.ylabel('NUMBER OF PEOPLE')
plt.show()
```

```python
# PIE CHART

fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
languages = ['English' , 'French' , 'Spanish' , 'Latin' , 'Green']
people = [100,50,150,40,70]
ax.pie(people, labels=languages , autopct='%1.1f%%' ) # autopct tells the floating value after decimal
plt.show()
```

```python
#Scatter Plot


x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)

fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1]) #2D form
ax.scatter(x,y , color = 'g')
ax.scatter(x,z , color = 'b')
plt.show()
```

```python
#3D Scatter Plot

fig3 = plt.figure()
ax = plt.axes(projection = '3d')
z = 20*np.random.random(100)
x = np.sin(z)
y = np.cos(z)
ax.scatter(x,y,c=z, cmap = 'Blues')
plt.show()
```

```python
#house price dataset 
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
house = pd.DataFrame(housing.data , columns = housing.feature_names)
house['PRICE'] = housing.target
```

```python
house.head()
```

Distrubution Plot

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])
```

Heat MAP

```python
correlation = house.corr()
```

```python
#constructing a heat map

plt.figure(figsize=(10,10))
sns.heatmap(correlation , cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
